```python
# ==========================
# Question 1
# Load the Loan Prediction dataset and perform initial analysis
# ==========================

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Loan prediction.csv")   # Change filename if required

# 1. Display first 10 records
print("========== First 10 Records ==========")
print(df.head(10))

# 2. Identify features and target variable
print("\n========== Features ==========")
print(list(df.columns[:-1]))

print("\n========== Target Variable ==========")
print(df.columns[-1])

# 3. Number of records and attributes
print("\n========== Dataset Shape ==========")
print("Number of Records:", df.shape[0])
print("Number of Attributes:", df.shape[1])

# 4. Check missing values
print("\n========== Missing Values ==========")
print(df.isnull().sum())

# 5. Statistical properties of numerical features
print("\n========== Statistical Summary ==========")
print(df.describe())
```

```python
# ==========================
# Question 2: Data Preprocessing
# ==========================

from sklearn.preprocessing import LabelEncoder, StandardScaler

# Create a copy of the dataset
data = df.copy()

# 1. Handle Missing Values
categorical_cols = ['Gender','Married','Dependents','Self_Employed','Loan_Amount_Term','Credit_History']

for col in categorical_cols:
    data[col].fillna(data[col].mode()[0], inplace=True)

data['LoanAmount'].fillna(data['LoanAmount'].median(), inplace=True)

# 2. Convert Categorical Variables into Numerical Format
le = LabelEncoder()

for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = le.fit_transform(data[col])

# 3. Create Feature Matrix (X) and Target Variable (y)
X = data.drop('Loan_Status', axis=1)
y = data['Loan_Status']

# 4. Standardize Numerical Features
scaler = StandardScaler()
X = scaler.fit_transform(X)

print("Data Preprocessing Completed Successfully!")

print("\nFeature Matrix Shape:", X.shape)
print("Target Variable Shape:", y.shape)

print("\nFirst 5 Rows of Processed Features:")
print(X[:5])

print("\nFirst 5 Target Values:")
print(y.head())
```

```python
# ==========================
# Question 3: Exploratory Data Analysis (EDA)
# ==========================

import matplotlib.pyplot as plt
import seaborn as sns

# Use original dataset for visualization
eda = df.copy()

# Handle missing values for plotting
eda['LoanAmount'].fillna(eda['LoanAmount'].median(), inplace=True)
eda['Credit_History'].fillna(eda['Credit_History'].mode()[0], inplace=True)
eda['Education'].fillna(eda['Education'].mode()[0], inplace=True)
eda['Property_Area'].fillna(eda['Property_Area'].mode()[0], inplace=True)

plt.figure(figsize=(6,4))
sns.countplot(x='Loan_Status', data=eda)
plt.title("Loan Approval Distribution")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x='Loan_Status', y='ApplicantIncome', data=eda)
plt.title("Applicant Income vs Loan Approval")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='Credit_History', hue='Loan_Status', data=eda)
plt.title("Credit History vs Loan Approval")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='Education', hue='Loan_Status', data=eda)
plt.title("Education Level vs Loan Approval")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='Property_Area', hue='Loan_Status', data=eda)
plt.title("Property Area vs Loan Approval")
plt.show()

print("----------- Observations -----------")
print("1. Credit History appears to be the most influential feature for loan approval.")
print("2. Applicants with good credit history have a much higher chance of loan approval.")
print("3. Higher applicant income generally increases the probability of loan approval.")
print("4. Graduates receive slightly more loan approvals than non-graduates.")
print("5. Semiurban areas have comparatively higher loan approval rates.")
```

```python
# ==========================
# Question 4
# Split Dataset & Train Models
# ==========================

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Train-Test Split (80:20)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# Models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42)
}

# Evaluation
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    results.append([name, accuracy, precision, recall, f1])

# Comparison Table
results_df = pd.DataFrame(results, columns=[
    "Model",
    "Accuracy",
    "Precision",
    "Recall",
    "F1 Score"
])

print("========== Model Comparison ==========")
print(results_df.round(4))
```

```python
# ==========================
# Question 6
# Stratified 5-Fold Cross Validation
# ==========================

from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

# Models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42)
}

# Stratified 5-Fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

results = []

for name, model in models.items():

    scores = cross_val_score(model, X, y, cv=skf, scoring='accuracy')

    print(f"\n{name}")
    print("Fold Accuracies:", np.round(scores,4))
    print("Mean Accuracy:", round(scores.mean(),4))
    print("Standard Deviation:", round(scores.std(),4))

    results.append([
        name,
        round(scores.mean(),4),
        round(scores.std(),4)
    ])

cv_results = pd.DataFrame(results,
                          columns=["Model",
                                   "Mean Accuracy",
                                   "Standard Deviation"])

print("\n========== Cross Validation Comparison ==========")
print(cv_results)
```

1. Why is Stratified K-Fold preferred for this dataset?

Answer:
Stratified K-Fold maintains the same proportion of approved and rejected loan applications in every fold. This produces a more reliable and unbiased evaluation, especially when the target classes are not perfectly balanced.

2. Which model shows the most consistent performance?

Answer:
The model with the lowest standard deviation in cross-validation scores is the most consistent. In most cases, Random Forest Classifier provides the most stable and consistent performance while maintaining high accuracy.

```python
# ==========================
# Question 7
# Hyperparameter Tuning using GridSearchCV
# ==========================

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Random Forest Model
rf = RandomForestClassifier(random_state=42)

# Parameter Grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 10],
    'min_samples_split': [2, 5, 10]
}

# GridSearchCV
grid = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

# Train GridSearch
grid.fit(X_train, y_train)

# Best Model
best_rf = grid.best_estimator_

# Predictions
y_pred = best_rf.predict(X_test)

# Evaluation
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("========== Hyperparameter Tuning Results ==========")

print("\nBest Parameters:")
print(grid.best_params_)

print("\nBest Cross Validation Score:")
print(round(grid.best_score_,4))

print("\nTest Set Performance")
print("Accuracy :", round(accuracy,4))
print("Precision:", round(precision,4))
print("Recall   :", round(recall,4))
print("F1 Score :", round(f1,4))

# Comparison Table
comparison = pd.DataFrame({
    "Model":["Random Forest (Before Tuning)",
             "Random Forest (After Tuning)"],
    "Accuracy":[
        results_df.loc[results_df["Model"]=="Random Forest","Accuracy"].values[0],
        accuracy
    ]
})

print("\n========== Before vs After Tuning ==========")
print(comparison)
```

1. Best Parameter Combination

Answer: The best parameter combination is the one displayed in the output under Best Parameters (it may vary slightly depending on the dataset).

2. Best Cross-Validation Score

Answer: The best cross-validation score is the value shown under Best Cross Validation Score.

3. Evaluation of Tuned Model

Answer: The tuned Random Forest model generally achieves better accuracy and overall performance compared to the default Random Forest model.

Comparison
Before Tuning: Uses default Random Forest parameters.
After Tuning: Uses the optimal parameters found by GridSearchCV, which usually improves model performance and generalization.

```python
# ==========================
# Question 8
# Bias-Variance Analysis using Decision Tree
# ==========================

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

depths = [2, 5, 15]
results = []

for depth in depths:

    model = DecisionTreeClassifier(max_depth=depth, random_state=42)

    model.fit(X_train, y_train)

    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    train_acc = accuracy_score(y_train, train_pred)
    test_acc = accuracy_score(y_test, test_pred)

    results.append([depth,
                    round(train_acc,4),
                    round(test_acc,4)])

# Create Result Table
bias_variance = pd.DataFrame(results,
                             columns=[
                                 "Max Depth",
                                 "Training Accuracy",
                                 "Testing Accuracy"
                             ])

print("========== Decision Tree Comparison ==========")
print(bias_variance)

print("\n========== Interpretation ==========")

print("Depth = 2  : Underfitting (High Bias)")
print("Depth = 5  : Balanced Bias and Variance")
print("Depth = 15 : Overfitting (High Variance)")
```

1. Which model exhibits underfitting?

Answer: The Decision Tree with max_depth = 2 exhibits underfitting because it is too simple to capture the data patterns, resulting in lower training and testing accuracy.

2. Which model exhibits overfitting?

Answer: The Decision Tree with max_depth = 15 exhibits overfitting because it fits the training data too closely, leading to very high training accuracy but comparatively lower testing accuracy.

3. Which model provides the best balance between bias and variance?

Answer: The Decision Tree with max_depth = 5 provides the best balance between bias and variance, as it generally achieves good training accuracy while maintaining strong testing accuracy.
